### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
# Enable Bokeh as backend for geoviews
gv.extension('bokeh')

# Load the CSV file with the city data
city_data_df = pd.read_csv('/Users/GURU/Desktop/python-api-challenge/output_data/cities.csv')

# Rename columns to use 'Latitude' and 'Longitude' instead of 'Lat' and 'Lng'
city_data_df.rename(columns={'Lat': 'Latitude', 'Lng': 'Longitude'}, inplace=True)

# Create a GeoViews Points object with humidity as the point size
points = gv.Points(city_data_df, kdims=['Longitude', 'Latitude'], vdims=['City', 'Country', 'Humidity'])

# Customize the points: size by humidity, add hover tools
points = points.opts(
    opts.Points(
        size=hv.dim('Humidity') * 0.1,  # Adjust multiplier for appropriate point size
        color='City',  # Use 'City' to assign different colors to each city
        cmap='Category20',  # Color map for colorful dots
        tools=['hover'],
        width=1200, height=600,
        xlabel='Longitude',
        ylabel='Latitude',
        bgcolor='lightblue'
    )
)

# Use OpenStreetMap tile source for the background
tiles = gv.tile_sources.OSM

# Combine the map tiles with the points
map_with_points = tiles * points

# Display the map
map_with_points
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python

# Load the CSV file with the city data
city_data_df = pd.read_csv('/Users/GURU/Desktop/python-api-challenge/output_data/cities.csv')

# Define your ideal weather conditions
min_temp = 20  # Minimum temperature in Celsius
max_temp = 30  # Maximum temperature in Celsius
min_humidity = 40  # Minimum humidity in percent
max_humidity = 70  # Maximum humidity in percent
max_cloudiness = 40  # Maximum cloudiness in percent
max_wind_speed = 10  # Maximum wind speed in km/h

# Narrow down cities that fit criteria
ideal_weather_df = city_data_df[
    (city_data_df['Max Temp'] >= min_temp) &
    (city_data_df['Max Temp'] <= max_temp) &
    (city_data_df['Humidity'] >= min_humidity) &
    (city_data_df['Humidity'] <= max_humidity) &
    (city_data_df['Cloudiness'] <= max_cloudiness) &
    (city_data_df['Wind Speed'] <= max_wind_speed)
]

# Drop any rows with null values
ideal_weather_df = ideal_weather_df.dropna()

# Display sample data
ideal_weather_df
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
# YOUR CODE HERE
hotel_df = ideal_weather_df[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
# YOUR CODE HERE
hotel_df['Hotel Name'] = ''

# Display sample data
# YOUR CODE HERE
hotel_df

```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python

# Set parameters to search for a hotel
radius = 10000  # Radius in meters
params = {
    'categories': 'accommodation.hotel',
    'limit': 1,
    'apiKey': '9f5f0eded7bd48eea7ad2f3f8de3c9b9'  # Replace with your actual Geoapify API key
}
#geoapify_key = "9f5f0eded7bd48eea7ad2f3f8de3c9b9"

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get latitude, longitude from the DataFrame
    lat = row['Lat']
    lng = row['Lng']

    # Add the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{lng},{lat},{radius}"
    params["bias"] = f"proximity:{lng},{lat}"

    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Make an API request using the params dictionary
    response = requests.get(base_url, params=params)

    # Convert the API response to JSON format
    name_address = response.json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found"
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df

```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
# Dependencies and Setup
import geoviews as gv
import holoviews as hv
import pandas as pd
from holoviews import opts
from geoviews import tile_sources as gts
from bokeh.io import output_notebook, show
from holoviews.plotting.bokeh import BokehRenderer

# Enable notebook output
output_notebook()

# Ensure the DataFrame columns have correct types and no missing values
assert 'Lat' in hotel_df.columns and 'Lng' in hotel_df.columns
assert hotel_df['Lat'].dtype in [float, int]
assert hotel_df['Lng'].dtype in [float, int]

# Check if the DataFrame is not empty
if not hotel_df.empty:
    # Create a GeoViews Points object with hotel name and country as hover information
    points = gv.Points(hotel_df, kdims=['Lng', 'Lat'], vdims=['Humidity', 'Hotel Name', 'Country', 'City'])

    # Customize the points: size, color, hover tools, and background
    points = points.opts(
        opts.Points(
            size=8,  # Fixed point size
            color='City',  # Use 'City' to assign different colors to each city
            cmap='Category20',  # Color map for colorful dots
            tools=['hover'],
            width=1200, height=600,
            xlabel='Longitude',
            ylabel='Latitude',
            bgcolor='lightblue'
        )
    )

    # Use OpenStreetMap tile source for the background
    tiles = gts.OSM

    # Combine the map tiles with the points
    map_with_points = tiles * points

    # Convert the Holoviews object to a Bokeh object
    bokeh_plot = hv.render(map_with_points, backend='bokeh')

    # Display the map using Bokeh show function
    show(bokeh_plot)
else:
    print("The DataFrame is empty. Please ensure the DataFrame has data before creating the map.")
```
